### 이름 : 법정동코드 dict 생성

```python
import pandas as pd

# 엑셀 파일 경로
file_path = r'C:\Users\user\dev\소스파일\FE\gongmo2\src\kmap\datalist.xlsx'

# 엑셀 파일 읽기
df = pd.read_excel(file_path)

# 법정동코드 , 법정동명 열
df = df[['법정동코드', '법정동명']]

# 법정동코드 dict 생성
code_name_dict = dict(zip(df['법정동명'], df['법정동코드']))
print(code_name_dict)

# json파일로 저장
import json

with open("sido_code_data.json", "w", encoding="utf-8") as f:
    json.dump(code_name_dict, f, ensure_ascii=False, indent=4)
```
